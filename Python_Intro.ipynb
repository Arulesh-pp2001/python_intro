###  Global Variables

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

```python
x = "awesome" #global variable

def myfunc():
  x = "fantastic" #local variable
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

### The global Keyword

```python
def myfunc():
  global x #treated as global variable
  x = "fantastic"

myfunc()

print("Python is " + x)
```

```python
x = "awesome" #global  variable

def myfunc():
  global x
  x = "fantastic" # changing global variable 

myfunc()

print("Python is " + x)
```
